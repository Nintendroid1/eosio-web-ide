```python
import numpy as np
import matplotlib.pyplot as plt
import math
import csv
import random

class patient:
  def __init__(self, name, gender, dob, address, phone, email, emergency, relationship, e_address, e_phone, e_email):
    self.name = '\"'+name+'\", '
    self.gender = '\"'+gender+'\", '
    self.dob =  '\"'+dob+'\",'
    self.address = '\"'+address+'\", '
    self.phone = '\"'+phone+'\", '
    self.email = '\"'+email+'\", '
    self.emergency = '\"'+emergency.replace("'","")+'\", '
    self.relationship = '\"'+relationship+'\", '
    self.e_address = '\"'+e_address+'\", '
    self.e_phone = '\"'+e_phone+'\", '
    self.e_email = '\"'+e_email+'\", '
    self.account = name.replace(" ", "").replace("'","").lower()[:12]
    
hospital = []

with open('MOCK_DATA.csv', newline='') as csvfile:
    reader = csv.reader(csvfile)
    for row in reader:
        p = patient(row[0],row[1],row[2],row[3],row[4],row[5],row[6],row[7],row[8],row[9],row[10])
#         print(p.name.replace(" ", "").lower()[:12])
        hospital.append(p)
    hospital = hospital[1:] #Skip header

KEY = "EOS75msrL9mzt8zT4btH6L7zzciaPQVQ1axpZmwkvB31qKm27nsNx"
with open('Account_Creation.sh', 'w') as f:
    f.write("#!/bin/bash\n")
    for p in hospital:
        f.write('cleos create account eosio ' + p.account + ' ' + KEY +'\n')

with open('insert_large.sh', 'w') as f:
    f.write("#!/bin/bash\n")
    for p in hospital:
        f.write("cleos push action hospital.vt addpatient \'[hospital.vt," +p.account+", " + p.gender + p.dob + p.address + p.phone + p.email + p.emergency + p.relationship + p.e_address + p.e_phone + p.email +" bob]\' -p hospital.vt -p bob\n")

with open('update_large.sh', 'w') as f:
    f.write("#!/bin/bash\n")
    for i in range(len(hospital)):
        f.write("cleos push action hospital.vt modvital \'[hospital.vt," +hospital[random.randint(0, len(hospital)-1)].account+"," +str(random.uniform(97, 101))+ ', '+str(random.uniform(60, 100))+ ', '+str(random.uniform(10, 17))+ ', '+str(random.uniform(80, 140))+"]\' -p hospital.vt \n")
    
with open('delete_large.sh', 'w') as f:
    f.write("#!/bin/bash\n")
    for i in range(len(hospital)):
        f.write("cleos push action hospital.vt delpatient \'[hospital.vt," +hospital[random.randint(0, len(hospital)-1)].account+"]\' -p hospital.vt \n")

with open('insert_large.sql', 'w') as f:
    for p in hospital:
        f.write("EXEC addpatient \"hospital.vt\", " + p.name + p.gender + p.dob + p.address + p.phone + p.email + p.emergency + p.relationship + p.e_address + p.e_phone + p.email +" \"bob\"\nGO\n")

with open('update_large.sql', 'w') as f:
    for i in range(len(hospital)):
        f.write("EXEC modvital \"hospital.vt\"," +hospital[random.randint(0, len(hospital)-1)].name+str(random.uniform(97, 101))+ ', '+str(random.uniform(60, 100))+ ', '+str(random.uniform(10, 17))+ ', '+str(random.uniform(80, 140))+"\nGO\n")
    
with open('delete_large.sql', 'w') as f:
    for i in range(len(hospital)):
        choice = hospital[random.randint(0, len(hospital)-1)].name
        f.write("EXEC modvital \"hospital.vt\"," +choice[:len(choice)-2]+"\nGO\n")
        
MEDIUM = 100
with open('insert_medium.sh', 'w') as f:
    f.write("#!/bin/bash\n")
    for p in hospital[:MEDIUM]:
        f.write("cleos push action hospital.vt addpatient \'[hospital.vt," +p.account+", " + p.gender + p.dob + p.address + p.phone + p.email + p.emergency + p.relationship + p.e_address + p.e_phone + p.email +" bob]\' -p hospital.vt -p bob\n")

with open('update_medium.sh', 'w') as f:
    f.write("#!/bin/bash\n")
    for i in range(MEDIUM):
        f.write("cleos push action hospital.vt modvital \'[hospital.vt," +hospital[random.randint(0, len(hospital)-1)].account+"," +str(random.uniform(97, 101))+ ', '+str(random.uniform(60, 100))+ ', '+str(random.uniform(10, 17))+ ', '+str(random.uniform(80, 140))+"]\' -p hospital.vt \n")
    
with open('delete_medium.sh', 'w') as f:
    f.write("#!/bin/bash\n")
    for i in range(MEDIUM):
        f.write("cleos push action hospital.vt delpatient \'[hospital.vt," +hospital[random.randint(0, len(hospital)-1)].account+"]\' -p hospital.vt \n")

with open('insert_medium.sql', 'w') as f:
    for p in hospital[:MEDIUM]:
        f.write("EXEC addpatient \"hospital.vt\", " + p.name + p.gender + p.dob + p.address + p.phone + p.email + p.emergency + p.relationship + p.e_address + p.e_phone + p.email +" \"bob\"\nGO\n")

with open('update_medium.sql', 'w') as f:
    for i in range(MEDIUM):
        f.write("EXEC modvital \"hospital.vt\"," +hospital[random.randint(0, len(hospital)-1)].name+str(random.uniform(97, 101))+ ', '+str(random.uniform(60, 100))+ ', '+str(random.uniform(10, 17))+ ', '+str(random.uniform(80, 140))+"\nGO\n")
    
with open('delete_medium.sql', 'w') as f:
    for i in range(MEDIUM):
        choice = hospital[random.randint(0, len(hospital)-1)].name
        f.write("EXEC modvital \"hospital.vt\"," +choice[:len(choice)-2]+"\nGO\n")
        

SMALL = 10
with open('insert_small.sh', 'w') as f:
    f.write("#!/bin/bash\n")
    for p in hospital[:SMALL]:
        f.write("cleos push action hospital.vt addpatient \'[hospital.vt," +p.account+", " + p.gender + p.dob + p.address + p.phone + p.email + p.emergency + p.relationship + p.e_address + p.e_phone + p.email +" bob]\' -p hospital.vt -p bob\n")

with open('update_small.sh', 'w') as f:
    f.write("#!/bin/bash\n")
    for i in range(SMALL):
        f.write("cleos push action hospital.vt modvital \'[hospital.vt," +hospital[random.randint(0, len(hospital)-1)].account+"," +str(random.uniform(97, 101))+ ', '+str(random.uniform(60, 100))+ ', '+str(random.uniform(10, 17))+ ', '+str(random.uniform(80, 140))+"]\' -p hospital.vt \n")
    
with open('delete_small.sh', 'w') as f:
    f.write("#!/bin/bash\n")
    for i in range(SMALL):
        f.write("cleos push action hospital.vt delpatient \'[hospital.vt," +hospital[random.randint(0, len(hospital)-1)].account+"]\' -p hospital.vt \n")

with open('insert_small.sql', 'w') as f:
    for p in hospital[:SMALL]:
        f.write("EXEC addpatient \"hospital.vt\", " + p.name + p.gender + p.dob + p.address + p.phone + p.email + p.emergency + p.relationship + p.e_address + p.e_phone + p.email +" \"bob\"\nGO\n")

with open('update_small.sql', 'w') as f:
    for i in range(SMALL):
        f.write("EXEC modvital \"hospital.vt\"," +hospital[random.randint(0, len(hospital)-1)].name+str(random.uniform(97, 101))+ ', '+str(random.uniform(60, 100))+ ', '+str(random.uniform(10, 17))+ ', '+str(random.uniform(80, 140))+"\nGO\n")
    
with open('delete_small.sql', 'w') as f:
    for i in range(SMALL):
        choice = hospital[random.randint(0, len(hospital)-1)].name
        f.write("EXEC modvital \"hospital.vt\"," +choice[:len(choice)-2]+"\nGO\n")

```
